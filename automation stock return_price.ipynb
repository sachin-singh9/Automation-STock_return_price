Create a function called,

    daily_return(df)
    
This function should be able to take a dataframe shown in the example above and calculate the daily return of the close price and output a dataframe with the same number of columns and an additional column named "**Return_Close**". 

The daily return for a day is calculated as follows:

    daily return = (today's price - previous day's price)/(previous day's price)*100

**Note:** You can calculate the daily return starting from the day 2. The return for the day 1 should be set to 0.

```python
#Defiing a function daily_return
def daily_return(df):
    #Taking the dataframe name and initialising it to a new variable
    dataframe=df
    #Creating an empty list 
    return_close=[]
    
    #For every row in the DataFrame calculating a new value as per the logic 
    #daily return = (today's price - previous day's price)/(previous day's price)*100
    for i in range(len(dataframe.Close)):
        #Excluding the first row
        if i>0:
            ret=((dataframe.Close[i]-dataframe.Close[i-1])/(dataframe.Close[i-1]))*100
            return_close.append(ret)
        else:
            #For first row intiliase it to 0
            return_close.append(0)
    #Adding a new column in DataFrame        
    dataframe["Return_Close"]=return_close    
    return(dataframe) #Return the newly created DataFrame
    
```

**Question 5**

Consider the NIFTY50 folder created in Q3. This folder should contain 50 data frames - one for each stocks in the Nifty-50 list. Create a function called,

    get_return(directory)
    

This function would take the link to the NIFTY50 folder, read one CSV file at a time present inside it, calulate the return of the close price, save the return value in an new column called "Return" and update the CSV file. 


```python
def get_return(directory=os.getcwd()):
    #opening folder nifty 50
    files=os.listdir(directory+"\\"+"NIFTY50")
    #iterating files in folder nifty 50
    for i in files:
        #assigning Return close empty column so that later on we can append the calculated return close price
        return_close=[]
        dframe=pd.read_csv(directory+"\\"+"NIFTY50"+ "\\" +i)    
        #iterating through close price for calculation
        for j in range(len(dframe.Close)):    
            if j>0:     #as we know first value is zero as we dont have previous data than starting date
                ret=((dframe.Close[j]-dframe.Close[j-1])/(dframe.Close[j-1]))*100
                return_close.append(ret)   #appended to empty list
            else:
                return_close.append(0)   #as first value is zero
        dframe["Return_Close"]=return_close  #adding column in dataframe
        dframe.to_csv(directory+"\\"+"NIFTY50"+ "\\" +i)  #upadating the files in the folder
```
